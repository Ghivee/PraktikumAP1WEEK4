```python
nilai = int(input('Masukkan Nilai ')) #baris 1

if(nilai >= 90):                      #baris 2
  print('Grade = A')                  #baris 3
elif(nilai >= 80):                    #baris 4
  print('Grade = B+')                 #baris 5
elif(nilai >= 70):                    #baris 6
  print('Grade = B')                  #baris 7
elif(nilai >= 60):                    #baris 8
  print('Grade = C+')                 #baris 9
elif(nilai >= 50):                    #baris 10
  print('Grade = C')                  #baris 11
elif(nilai >= 40):                    #baris 12
  print('Grade = D')                  #baris 13
else:                                 #baris 14
  print('Grade = E')                  #baris 15
```

```python
nilai = float(input('Masukkan Nilai '))

if  (80.00 <= nilai <= 100):
  print('Grade = A')
elif(77.50 <= nilai <= 79.99):
  print('Grade = A-')
elif(75.00 <= nilai <= 77.49):
  print('Grade = A/B')
elif(72.50 <= nilai <= 74.99):
  print('Grade = B+')
elif(70.00 <= nilai <= 72.49):
  print('Grade = B')
elif(67.50 <= nilai <= 69.99):
  print('Grade = B-')
elif(65.00 <= nilai <= 67.99):
  print('Grade = B/C')
elif(62.50 <= nilai <= 64.99):
  print('Grade = C+')
elif(60.00 <= nilai <= 62.49):
  print('Grade = C')
elif(55.00 <= nilai <= 59.99):
  print('Grade = C-')
elif(50.00 <= nilai <= 54.99):
  print('Grade = C/D')
elif(45.00 <= nilai <= 49.99):
  print('Grade = D+')
elif(40.00 <= nilai <= 44.99):
  print('Grade = D')
else:
  print('Grade = E')
```
